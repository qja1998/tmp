```python
features = optimized_route['features']
len(features)
```

```python
line_strings = []
for feature in features:
    _type = feature['type']
    _geometry = feature['geometry']
    _type_of_geometry = _geometry['type']
    print(_type_of_geometry)
    if _type_of_geometry == "LineString":
        line_strings.append(_geometry['coordinates'])
print(len(line_strings))
```

```python
coordinates = []
num_of_coordinates = 0
for line_string in line_strings:
    num_of_coordinates += len(line_string)
    coordinates.extend(line_string)
    print(len(line_string))
print(num_of_coordinates)
```
